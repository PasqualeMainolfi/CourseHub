```python
VAR = ["W", "H", "Q", "E", "S"]
CONST = ""
RULES = { "W": "QQQEE", "H": "EQSS", "Q": "WH", "S": "H", "E": "Q" } # A -> AB, B -> A
ASSIOMA = "H"
```

```python
lsys = ASSIOMA

t = 0
tmax = 10

seqs = []

while t < tmax:
    
    temp = ""
    for var in lsys:
        if var in RULES:
            temp += RULES[var]
        
    print(f"[n = {t}] = {lsys}")
    seqs.append(lsys)
    
    lsys = temp
    t += 1
```

```python
bpm = 250
beats = { "W": bpm / 4, "H": bpm / 2, "Q": bpm, "E": bpm * 2, "S": bpm * 4 }

start_time = 0
for b in seqs[5]:
    dur = 60 / beats[b]
    score = f"i 1 {start_time} {dur}"
    print(score)
    start_time += dur
    

```
